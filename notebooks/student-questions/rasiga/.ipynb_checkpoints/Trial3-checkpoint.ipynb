```python
# Blocking
## plan : 
### A, B -----------overlap blocking (title)-----------------------------> candset  |
### A, B ------------rule-based-blocking(title & director)----------------> candset |----union--->candset

import magellan as mg
A = mg.read_csv('tableA.csv', key='ID')
B = mg.read_csv('tableB.csv', key='ID')
ob = mg.OverlapBlocker()
ab = mg.AttrEquivalenceBlocker()
```

```python
C = ab.block_tables(A, B, 'Title', 'Title', 'Title', 'Title')
C.head()
```

```python
D = ob.block_tables(A, B, 'Title', 'Title', l_output_attrs=['Title'], r_output_attrs=['Title'])
D
```

```python
feature_table = mg.get_features_for_blocking(A, B)
feature_table
```

```python
rb = mg.RuleBasedBlocker()
rb.add_rule(['Title_Title_lev(ltuple,rtuple) < 0.8','Director_Director_nmw(ltuple,rtuple) < 0.8'], feature_table)
```

```python
mg.init_jvm() #Do we even need this? Doesn't work on my machine.
```

```python
#WORKING PERFECTLY! EDIT DISTANCE!
E = rb.block_tables(A, B, l_output_attrs=['Title'], r_output_attrs=['Title'])
E
```

```python
def year_blocker_function(x, y):
    x_date = str(x['Year'])
    y_date = str(y['Year'])
    if(x_date=='NULL' or y_date=='NULL' or x_date == 'nan' or y_date == 'nan'):
        return False #Since one of them is NULL, we cannot block using this feature
    x_year = int(x_date.split('-')[0])
    y_year = int(y_date.split('-')[0])
    
    if(x_year > y_year + 1 or x_year < y_year - 1):
        return True #Block the tuples
    
    return False

bb = mg.BlackBoxBlocker()
bb.set_black_box_function(year_blocker_function)
```

```python
F = bb.block_tables(A, B, l_output_attrs=['Title'], r_output_attrs=['Title'])
```

```python
F #NOW F will have all movie pairs that have been released within one year of each other
#We need to apply this along with other blockers to make it practical
```
